```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

# Mengatasi Data Hilang (Preprocessing)

```python
#menghapus baris yang memiliki data yang hilang/rusak
print(data['ammonia'].value_counts()['#NUM!'], 'data yang rusak') #Menghitung berapa banyak data rusak pada kolom amonia
data = data[data['ammonia'].str.contains('#NUM!') == False] #Menghapus baris pada kolom amonia yang rusak
```

```python
#mengubah column dari object menjadi data type numerik (integer)
data['ammonia'] = pd.to_numeric(data['ammonia'])
data['is_safe'] = pd.to_numeric(data['is_safe'])
```

**Menentukan input dan output**
<br>
Menyeimbangkan target variable dan menyimpan nilai input ke dalam variabel X dan nilai output akan disimpan di dalam variabel y. Seperti yang dapat kita lihat pada grafik di atas, variabel target sangat tidak seimbang maka kita harus menanganinya. Kami menggunakan strategi oversampling dengan menggunakan kelas RandomOverSampler, sehingga kami akan menghasilkan jumlah baris yang sama untuk nilai 1 dari variabel target "is_safe". Jika kami tidak menangani variabel target yang tidak seimbang ini, maka model pembelajaran mesin akan condong ke kelas 0 daripada 1.

```python
import seaborn as sns
palette_features = ['#E68753', '#409996']

#plot target variable
sns.set(rc={'axes.facecolor':'#ECECEC'}) #background color of plot
plt.figure(figsize=(12,6))
plt.title("Imbalanced target variable", fontsize=15, ha='center')
ax = sns.countplot(x=data['is_safe'], data=data, palette=palette_features)

#show the plot
plt.show()
```

```python
from imblearn.over_sampling import RandomOverSampler
from collections import Counter

#balance the target variable
X = data.drop(['is_safe'],axis=1)
y = data['is_safe']

ros = RandomOverSampler(sampling_strategy='minority') #RandomOverSampler akan menghasilkan kembali data yang seimbang dengan menambahkan duplikat pada kategori minoritas
X, y = ros.fit_resample(X, y) #y_train as balanced target variable
print(f"Imbalanced target class:\n{(y)}\n\nBalanced target class: {Counter(y)}\n")
print(X.shape[0] - data.shape[0], 'new random picked points')
```

```python
#plot the balanced target variable
sns.set(rc={'axes.facecolor':'#ECECEC'}) #background color of plot
plt.figure(figsize=(12,6))
plt.title("Balanced target variable", fontsize=15, ha='center')
ax = sns.countplot(x=y, data=data, palette=palette_features)

plt.show()
```

**Membagi data latih dan data uji**
<br>
Setelah diperolah nilai input & output di dalam variabel X & y, selanjutnya kita akan membagi data menjadi data latih dan data uji dengan proporsi sebagai berikut: 80% data latih dan 20% data uji.

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
display(X_train)
display(y_train)
```

**Feature Scaling**
<br>
Untuk memastikan angka di dalam satu kolom/fitur tidak memiliki jarak nilai yang terlalu besar atau kecil baik untuk nilai lain yang berada di dalam kolom yang sama ataupun kolom yang lain, kita bisa menggunakan StandardScaler untuk melakukan normalisasi. Pada dasarnya setelah melakukan normalisasi semua nilai akan berada di dalam range nilai yang kurang lebih sama. Rumusnya adalah sebagai berikut:
<br>
\begin{equation}
z = \frac{(x-u)}{s} \tag{1}
\end{equation}
<br>
Dengan:
* x = nilai pada baris dan kolom tertentu
* u = nilai rata-rata kolom tertentu
* s = nilai standar deviasi kolom tertentu

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

## Decision Tree Model

**Membuat model machine learning: Decision Tree**

Setelah kita menerapkan StandardScaler pada nilai X_train dan X_test, sekarang kita akan memasukkan data tersebut ke dalam model machine learning untuk klasifikasi. Adapun model yang digunakan adalah Decision Tree yang pada prinsipnya adalah sekumpulan logika if-else yang digunakan untuk melakukan klasifikasi.

**Kelebihan:**
- Bisa digunakan pada data yang sudah di scaling ataupun belum
- Data yang hilang tidak ada mempengaruhi pembentukan pohon keputusan
- Sangat intuitif dan mudah dimengerti karena terdapat visualisi pohon

**Kekurangan:**
- Perubahan kecil pada data dapat menyebabkan perubahan besar pada struktur pohon keputusan yang menyebabkan ketidakstabilan
- Seringkali melibatkan waktu yang lebih lama untuk melatih data
- Terkadang perhitungan bisa jauh lebih kompleks dibandingkan dengan algoritma lain



```python
# Decision Tree classifier
from sklearn.tree import DecisionTreeClassifier
# visulasi
classifier = DecisionTreeClassifier(max_depth = 2)
classifier.fit(X_train, y_train)
```

```python
# Prediksi nilai y
y_pred_dt = classifier.predict(X_test)
y_pred_dt
```

```python
data.is_safe.unique()
```

**Visualisasi Decision Tree**
<br>
Model Decision Tree yang sudah kita buat dapat ditampilkan dengan menggunakan plot_tree yang terdapat di dalam library sklearn.tree

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (3,2), dpi=300)
fn = data.columns[0:20]
cn = ['Not Safe', 'Safe']
plot_tree(classifier, feature_names = fn, class_names=cn, filled = True);
```

```python
# Tuliskan label dari masing-masing kelas sesuai urutan
class_names = data["is_safe"].unique()
class_names = np.sort(class_names)
class_names
```

**Confusion Matrix**
Untuk mengetahui tingkat akurasi dari model yang sudah dibuat dapat dilakukan dengan menggunakan confusion matrix. 

```python
# Making the Confusion Matrix
from sklearn import metrics

cm = metrics.confusion_matrix(y_test, y_pred_dt)

# Membangun heatmap
fig, ax = plt.subplots()
sns.heatmap(pd.DataFrame(cm), annot=True, cmap="YlGnBu" ,fmt='g')
ax.xaxis.set_label_position("top")
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names, ha="center")
plt.yticks(tick_marks, class_names, rotation='horizontal')
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred_dt))
```

## Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression

# Define the classifier
logreg = LogisticRegression()

# Performing the model building
logreg.fit(X_train,y_train)

# Predicting the labels of the test dataset
y_pred_lr=logreg.predict(X_test)

# Comparing the true labels vs predicted labels
compare = pd.DataFrame({"y_true": y_test, "y_pred": y_pred_lr})
display(compare)
```

```python
# Fungsi untuk menghitung confusion matrix
cnf_matrix = metrics.confusion_matrix(y_test, y_pred_lr, labels = class_names)

### Berikut merupakan kode untuk memproduksi heatmap visualisasi confusion matrix
#Confusion matrix adalah tabel yang digunakan untuk mengevaluasi kinerja model 
#dengan menghitung jumlah True Positive (TP), False Positive (FP), True Negative (TN), dan False Negative (FN).

# Membangun heatmap
fig, ax = plt.subplots()
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
ax.xaxis.set_label_position("top")
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names, ha="center")
plt.yticks(tick_marks, class_names, rotation='horizontal')
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

```python
print(classification_report(y_test, y_pred_lr))
```

## K-Nearest Neighbors Model

**Criteria K-NN Model:**
- Address Missing Data
- Rescale Data
- Reduce Dimensionality

**Kelebihan:**
- Tidak perlu training sebelum melakukan prediksi
- Mempersingkat waktu
- Mudah untuk diimplementasi

**Kekurangan:**
- Kurang maksimal ketika digunakan pada data yang berjumlah sangat besar
- Sensitif terhadap data hilang, rusak, dan data pencilan
- Memerlukan feature scaling sebelum dapat diterapkan


**Reduce Dimensionality**

Reduce Dimensionality dilakukan dengan menggunakan salah satu metode Seleksi Fitur. Seleksi fitur adalah konsep penting dalam ML. Konsep ini membantu meningkatkan kekuatan prediksi algoritma dengan memilih fitur yang paling kritis untuk memprediksi variabel target dan menghilangkan yang redundan (pengurangan dimensi).

```python
from sklearn.feature_selection import SelectKBest, f_classif

# mencari 10 fitur terbaik
select_feature = SelectKBest(f_classif, k=10).fit(X_train, y_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier

# mengimplementasi algoritma K-NN
knnc = KNeighborsClassifier() 
#algoritma supervised learning yang digunakan untuk melakukan klasifikasi data berdasarkan k-NN dari setiap titik pada data
knnc.fit(X_train, y_train) # melakukan pelatihan model KNN
y_pred_knnc = knnc.predict(X_test) #melakukan prediksi pada data uji dengan menggunakan model KNN yang telah dilatih

# Fungsi untuk menghitung confusion matrix
cnf_matrix_knnc = metrics.confusion_matrix(y_test, y_pred_knnc, labels = class_names)

### Berikut merupakan kode untuk memproduksi heatmap visualisasi confusion matrix

# Membangun heatmap
#membandingkan nilai prediksi dari model dengan nilai yang sebenarnya
fig, ax = plt.subplots()
sns.heatmap(pd.DataFrame(cnf_matrix_knnc), annot=True, cmap="YlGnBu" ,fmt='g')
ax.xaxis.set_label_position("top")
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names, ha="center")
plt.yticks(tick_marks, class_names, rotation='horizontal')
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

```python
#evaluation of K-NN: metrics pivot chart
print(classification_report(y_test, y_pred_knnc))
```

# Kesimpulan

Accuracy = menunjukkan seberapa akurat model dalam memprediksi kelas target.

Precision = menunjukkan seberapa akurat model dalam memprediksi positif yang benar (true positive).

Recall = menunjukkan seberapa banyak dari total kelas positif yang model berhasil memprediksi.

F-Score adalah harmonic mean dari precision dan recall, dan merupakan metrik yang baik untuk mengevaluasi kinerja model pada dataset yang tidak seimbang (imbalanced dataset). Semakin tinggi skor F1, semakin baik kinerja model.

```python
print(classification_report(y_test, y_pred_dt))
```

```python
print(classification_report(y_test, y_pred_lr))
```

```python
print(classification_report(y_test, y_pred_knnc))
```

Dapat disimpulkan bahwa model prediksi water quality menggunakan Decision Tree Classifier (DTC), Logistic Regression (LR), maupun K-Nearest Neighbors Classifier (KNNC) telah berjalan dengan sangat baik, karena hasil evaluasi model tersebut melalui output dari classification_report,baik dari precision, recall, dan f1-scorenya menunjukan nilai secara rata-rata adalah diatas 0.8. Selain itu, akurasi keseluruhan model juga diberikan melalui output dari accuracy_score. Nilai akurasi ini memberikan informasi mengenai seberapa baik model dalam memprediksi kelas secara keseluruhan.

Dalam kesimpulannya, dapat dikatakan bahwa model prediksi water quality menggunakan K-Nearest Neighbors Classifier (KNNC) merupakan model yang terbaik setelah Decision Tree.
